```python
import cv2
import os
```

```python
# Opening a video file
cam = cv2.VideoCapture('video.mp4')
```

```python
# Making a folder datanew in the same directory

try:  
    if not os.path.exists('datanew'): 
        os.makedirs('datanew') 
except OSError: 
    print ('Error: Creating directory of data') 
```

```python
# Capturing the frames from the video
currentframe = 0
  
while(True):  
    ret,frame = cam.read() 
  
    if ret: 
        name = './datanew/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name)  
        cv2.imwrite(name, frame) 
   
        currentframe += 1
    else: 
        break
        
cam.release() 
cv2.destroyAllWindows() 
```
